```python
def convert_area(value):
  if 'м²' in value:
    value = value.replace('м²', '')
    value = value.strip()
    if ' ' in value:
      value = value.replace(' ', '.')
      return float(value)
    else:
      return float(value)
  elif ' ' in value:
    value = value.replace(' ', '.')
    return float(value)
  else:
    value = value.strip()
    return float(value)
df['areas'] = df['areas'].map(convert_area)
df['areas'].value_counts()
```

```python
def room(value):
  if '+' in value:
    value = value.replace('+', '')
    return float(value)
  elif value == 'Свободная планировка':
    value = value.replace('Свободная планировка', '')
    return float(0)
  else:
    return float(value)
df['rooms'] = df['rooms'].map(room)
df['rooms'].value_counts()
```

```python
house_floor = []
floor_th = []
def floor_h(value):
  if '/' in value:
    splt = value.split("/")
    floor_th.append(float(splt[0]))
    house_floor.append(float(splt[1]))
    return value

df['floors'] = df['floors'].apply(floor_h)
# df['floors'].value_counts()
print(len(house_floor))
print(len(floor_th))
```

```python
df['house_floor'] = house_floor
df['floor_th'] = floor_th
```

```python
df.drop(['floors'], axis=1, inplace=True)
```

```python
df['renovations'].value_counts()
```

```python
def price(value):
  if 'у.е.' in value:
    value = value.replace('у.е.','')
    value = value.strip()
    if ' ' in value:
      value = value.replace(' ','')
      return float(value)
    else:
      return float(value)
  elif value == 'No price':
    return float(0)
  else:
    return float(value)
df['prices'] = df['prices'].map(price)
df['prices'].value_counts()
```

```python
def add(value):
  if 'Ташкент' in value:
    res = value.split(" ")
    total = res[2:4]
    text = ""
    for i in total:
      text += i
    return text
  elif 'область' in value:
    res = value.split(" ")
    total = res[0:4]
    text = ""
    for i in total:
      text += i
    return text
  else:
    return 'Юнусабадскийрайон,'
df['addresses'] = df['addresses'].apply(add)
df['addresses'].value_counts()
```

```python
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import mean_squared_error
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
import torch
from torch import nn
from torch.optim import Adam
```

```python
x = df.drop('prices', axis=1)
y = df['prices']
```

```python
sc = StandardScaler()
oe = OneHotEncoder()
```

```python
numeric = df.select_dtypes(include='float64').columns
categoric = df.select_dtypes(include='object').columns
print(numeric)
print(categoric)
```

```python
num_transformer = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('num',sc)
    ]
)
cat_transformer = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('cut',oe)
    ]
)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer,numeric.drop('prices')),
        ('cut', cat_transformer,categoric)
    ]
)
print(preprocessor)
```

```python
x_processed = preprocessor.fit_transform(x)
y_processed = np.log1p(y)
```

```python
x_train, x_test, y_train, y_test, = train_test_split(x_processed,y_processed, test_size=0.25, random_state=42)
```

```python
print(x_train)
```

```python
x_train_dense = x_train.toarray()
x_test_dense = x_test.toarray()
x_train_tensor = torch.FloatTensor(x_train_dense)
x_test_tensor = torch.FloatTensor(x_test_dense)
y_train_tensor = torch.FloatTensor(y_train.values).reshape(-1,1)
y_test_tensor = torch.FloatTensor(y_test.values).reshape(-1,1)
```

```python
print(f"{x_train_tensor}")
print(f"{x_test_tensor}")
print(f"{y_train_tensor}")
print(f"{y_test_tensor}")
```

```python
class HousePrice(nn.Module):
  def __init__(self, input_size):
    super().__init__()
    self.layer1 = nn.Linear(input_size, 40)
    self.layer2 = nn.Linear(40, 10)
    self.layer3 = nn.Linear(10, 5)
    self.layer4 = nn.Linear(5, 1)
    self.dropout = nn.Dropout(0.2)

  def forward(self, x):
    x = torch.relu(self.layer1(x))
    x = self.dropout(x)
    x = torch.relu(self.layer2(x))
    x = self.dropout(x)
    x = torch.relu(self.layer3(x))
    x = self.dropout(x)
    x = self.layer4(x)
    return x

input_size = x_train_tensor.shape[1]
model = HousePrice(input_size)
criterion = nn.MSELoss()
optimizer = Adam(model.parameters(), lr=0.001)
epoch = 200
batch_size = 50
train_losses = []
test_losses = []

for i in range(epoch):
  model.train()
  epoch_loss = 0
  for j in range(0, len(x_train_tensor), batch_size):
    batch_x = x_train_tensor[j:j+batch_size]
    batch_y = y_train_tensor[j:j+batch_size]
    outputs = model(batch_x)
    loss = criterion(outputs,batch_y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    epoch_loss += loss.item()
  avg_loss = epoch_loss/len(x_train_tensor/batch_size)
  train_losses.append(avg_loss)

  model.eval()
  with torch.no_grad():
    test_outputs = model(x_test_tensor)
    test_loss = criterion(test_outputs,y_test_tensor)
    test_losses.append(test_loss.item())

  if (i + 1) % 20 == 0:
    print(f"Epoch {i+1}/{epoch} Train Loss: {avg_loss}. Test loss: {test_loss.item()}")
```

```python
model.eval()
with torch.no_grad():
  test_outputs = model(x_test_tensor)
  test_loss = criterion(test_outputs,y_test_tensor)
  test_mse = mean_squared_error(y_test_tensor.numpy(), test_outputs.numpy())
  print(f"Test mse: {np.sqrt(test_mse)}")
```
